```python
%matplotlib inline  
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import math
import statsmodels.api as sm
import csv
```

```python
data = np.load('9-1.npy')
temp = st.expon.rvs(loc=0, scale=1/(1.0), size=len(data), random_state=None)
#data = st.expon.rvs(loc=0, scale=1/(0.9), size=1000, random_state=None)
print(1/np.mean(data))
print(1/np.mean(temp))
```

В теоретических задачах были получены два равноменрно наиболее мощных критерия $S$ уровня значения $\alpha$ проверки гипотезы
- $H_0$: $\theta = \theta_0$ vs $H1$: $\theta > \theta_0$
- $H_0$: $\theta = \theta_0$ vs $H1$: $\theta < \theta_0$

Для первого $S = \{ \sum{X_i} < z_{\alpha} \} $ 

Для второго $S = \{ \sum{X_i} > z_{1 -\alpha} \}$

```python
tetas = np.array([0.9,1,1.1])
```

Создадим критерий для каждого предполагаемого значения $\theta ={0.9,1.0,1.1}$ для различных уровней значимости.

```python
set1 = sum(data)
set2 = sum(temp)
n = len(data)
c_results = []
b_results = []
a_results = []
c_results1 = []
b_results1 = []
a_results1 = []
level = np.zeros(3)
level1 = np.zeros(3)
for i in range(2, 100):
    for j in range(3):
        level[j] = st.gamma.ppf(1-1/i, n, loc=0, scale=1/tetas[j])
        level1[j] = st.gamma.ppf(1/i, n, loc=0, scale=1/tetas[j])
    a_results.append(level[0])  
    b_results.append(level[1])
    c_results.append(level[2])
    a_results1.append(level1[0])  
    b_results1.append(level1[1])
    c_results1.append(level1[2])
```

Подберем такой уровень значимости, чтобы 

```python
plt.plot(range(2, 100), c_results, label = '1.1', color='green')
plt.plot(range(2, 100), b_results, label = '1.0', color ='blue')
plt.plot(range(2, 100), a_results, label = '0.9', color='black')
plt.plot(range(2, 100), c_results1, label = '1.1', color='green')
plt.plot(range(2, 100), b_results1, label = '1.0', color='blue')
plt.plot(range(2, 100), a_results1, label = '0.9', color='black')
plt.plot(range(2, 100), set1*np.ones(98), label = 'data')
plt.plot(range(2, 100), set2*np.ones(98), label = 'temp')
fig = plt.gcf()
plt.grid(True)
fig.set_size_inches(18.5, 10.5)
plt.legend( fontsize=13)
plt.show()
```

```python
alpha = np.argmin(b_results + np.ones(98)*set1)
print(alpha)
```

Таким образом видно, что истинное значение $\theta$ равно $0.9$ с уравнем значения $\alpha$ уровень значимости статистического критерия – это ограничение сверху на вероятность ошибки первого рода (вероятность отвергнуть нулевую гипотезу, когда она на самом деле верна).
