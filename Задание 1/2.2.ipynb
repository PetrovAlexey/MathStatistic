```python
%matplotlib inline  
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import math
```

–ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫—É –∏–∑ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º $\theta = 1$ —Ä–∞–∑–º–µ—Ä–∞ $N = 10000$

```python
teta = 1.0 #–ø–∞—Ä–∞–º–µ—Ç—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–Ω–∏—è
expon = st.expon(loc=0, scale=1.0/teta)
a = expon.rvs(size=10000)
```

–î–ª—è –∫–∞–∂–¥–æ–≥–æ $n < N$ —Å—á–∏—Ç–∞–µ–º –æ—Ü–µ–Ω–∫—É $\left(\frac{ùëò!}{\overline{ùëã^{ùëò}}}\right)Ô∏Å^ {(1/ùëò)}$ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\theta$ = 1.
(–¥–µ–ª–∞—é –Ω–µ –≤ —Ü–∏–∫–ª–µ –ø–æ—Ç–æ–º—É, —á—Ç–æ –æ—á–µ–Ω—å –¥–æ–ª–≥–æ —Å—á–∏—Ç–∞–µ—Ç)

```python
k=1

fact = math.factorial(k)
stat2 = np.zeros(10001)
for i in range(1,10001):
    m_k = [(z)**(k) for z in a[:i]]
    mom_k = np.sum(m_k)/i
    stat2[i] = math.pow((fact/mom_k),(1.0/k))
print ("–ó–Ω–∞—á–µ–Ω–∏–µ –æ—Ü–µ–Ω–∫–∏ –Ω–∞ –≤—Å–µ–π –≤—ã–±–æ–∫–µ: " + str(stat2[10000]))
```

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –ø—Ä–∏ $k=2,3$. –ß—Ç–æ–±—ã —É–±–µ–¥–∏—Ç—å—Å—è, —á—Ç–æ –≤—Å–µ –¥–µ–ª–∞–µ–º –ø—Ä–∞–≤–∏–ª—å–Ω–æ.

```python
plt.plot(np.arange(1,10001), abs(stat1 - teta)[1:], label =r'k = 3')
plt.plot(np.arange(1,10001), abs(stat0 - teta)[1:], label =r'k = 2')
plt.xlabel(r'$x$') #–ú–µ—Ç–∫–∞ –ø–æ –æ—Å–∏ x –≤ —Ñ–æ—Ä–º–∞—Ç–µ TeX
plt.ylabel(r'$f(x)$') #–ú–µ—Ç–∫–∞ –ø–æ –æ—Å–∏ y –≤ —Ñ–æ—Ä–º–∞—Ç–µ TeX
plt.legend(fontsize=12, loc=1)
plt.grid(True) #–°–µ—Ç–∫–∞
fig = plt.gcf()
plt.ylim (0, 0.2)
fig.set_size_inches(18.5, 10.5)
plt.show()
```
