```python
%matplotlib inline  
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import math
import statsmodels.api as sm
```

```python
g = 1.0 #параметр распредения
a = 0.0
norm = st.norm(loc=a, scale=g)
sample = norm.rvs(size=1000)
```

```python
stat1_a = np.zeros(1001)
stat1_g = np.zeros(1001)
for i in range(1,1001):
    stat1_a[i] = np.mean(sample[:i])
    m_k = [(z)**(2) for z in sample[:i]]
    mom_k = np.sum(m_k)/i
    stat1_g[i] = mom_k - stat1_a[i]**2
print (stat1_g[1000])
print (stat1_a[1000])
```

Параметрический бутстрап

```python
Par = 1000

boot_stat1_a = np.zeros((1001,1001))
boot_stat1_g = np.zeros((1001,1001))
disp_boot_stat1_a = np.zeros(1001)
disp_boot_stat1_g = np.zeros(1001)
for i in range(1,1001):
    boot_norm = st.norm(loc=stat1_a[i], scale=stat1_g[i])
    boot_sample = np.zeros((1001,i))
    for j in range(1,Par+1):
        boot_sample[j] = boot_norm.rvs(size=i)
        boot_stat1_a[i][j] = np.mean(boot_sample[j])
        #m_k = [(z)**(2) for z in boot_sample[j]]
        #mom_k = np.sum(m_k)/Par
        boot_stat1_g[i][j] = st.moment(boot_sample[j], 2)
        #boot_stat1_g[i][j] = mom_k - boot_stat1_a[i][j]**2

    disp_boot_stat1_a[i] = st.moment(boot_stat1_a[i], 2)
    disp_boot_stat1_g[i] = st.moment(boot_stat1_g[i], 2)
    #disp_boot_stat1_g[i] = smom_kg - np.mean(boot_stat1_g[i])

print(disp_boot_stat1_a[1000])

print(disp_boot_stat1_g[1000])
```

```python
plt.subplot (1, 2, 1)
plt.plot(np.arange(1,1001), disp_boot_stat1_a[1:])
plt.xlabel(r'$n$') #Метка по оси x в формате TeX
plt.ylabel(r'$s(\theta)$') #Метка по оси y в формате TeX
#plt.title(r'$y=x^2$') #Заголовок в формате TeX
plt.grid(True) #Сетка
plt.subplot (1, 2, 2)
plt.plot(np.arange(1,1001), disp_boot_stat1_g[1:])
plt.xlabel(r'$n$') #Метка по оси x в формате TeX
plt.ylabel(r'$s(\theta)$') #Метка по оси y в формате TeX
#plt.title(r'$y=x^2$') #Заголовок в формате TeX
plt.grid(True) #Сетка
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

```

Непараметрический бутстрап

```python
Par = 1000

npboot_stat1_a = np.zeros((1001,1001))
npboot_stat1_g = np.zeros((1001,1001))
npdisp_boot_stat1_a = np.zeros(1001)
npdisp_boot_stat1_g = np.zeros(1001)
for i in range(1, 1001):
    npboot_sample = np.zeros((Par+1, i))
    for j in range(1, Par+1):
        x = np.random.randint(len(sample[:i]), size = len(sample[:i]))
        npboot_sample[j] = np.take(sample[:i], x)
        npboot_stat1_a[i][j] = np.mean(npboot_sample[j])
        npboot_stat1_g[i][j] = st.moment(npboot_sample[j], 2)
    npdisp_boot_stat1_a[i] = st.moment(npboot_stat1_a[i], 2)
    npdisp_boot_stat1_g[i] = st.moment(npboot_stat1_g[i], 2)

print(npdisp_boot_stat1_a)

print(npdisp_boot_stat1_g)
```

```python
plt.subplot (1, 2, 1)
plt.plot(np.arange(1,1001), npdisp_boot_stat1_a[1:])
plt.xlabel(r'$n$') #Метка по оси x в формате TeX
plt.ylabel(r'$s(\theta)$') #Метка по оси y в формате TeX
#plt.title(r'$y=x^2$') #Заголовок в формате TeX
plt.grid(True) #Сетка
plt.subplot (1, 2, 2)
plt.plot(np.arange(1,1001), npdisp_boot_stat1_g[1:])
plt.xlabel(r'$n$') #Метка по оси x в формате TeX
plt.ylabel(r'$s(\theta)$') #Метка по оси y в формате TeX
#plt.title(r'$y=x^2$') #Заголовок в формате TeX
plt.grid(True) #Сетка
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

```
